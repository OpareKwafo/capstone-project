```python
#2022 ->2020 season
#predict who will win a match

import requests 
standings_url = "https://fbref.com/en/comps/12/La-Liga-Stats"
```

```python
data = requests.get(standings_url)
#data.text -> view raw data
#we want the url for reach squad from the league table. go to each page and grab the match log.
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data.text)   #initialise our soup
```

```python
standings_table = soup.select('table.stats_table')[0]       #returns a list.need the first one.
#standings_table #only html table
```

```python
# grabbing all the 'a' tags inside the table above
# select uses css selectors for classes ids etc, find all find tags
links = standings_table.find_all('a')
#can make this a list comprehension
links = [l.get("href") for l in links]
```

```python
#filtering the link
links = [l for l in links if '/squads' in l]
```

```python
links
```

```python
#turning links into full urls ie absolute links
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

```python
#extracting match stats
team_url = team_urls[0]
data = requests.get(team_url)
```

```python
import pandas as pd
import html5lib
#match looks for a specific string inside the table and read_html reads all tables on the page
matches = pd.read_html(data.text, match="Scores & Fixtures")
matches[0]  #returns a list. grabbing the first one
```

```python
#grabbing the shooting stats
soup = BeautifulSoup(data.text)
links = soup.find_all('a')
links = [l.get("href") for l in links]
links = [l for l in links if l and 'all_comps/shooting' in l]  #link to shooting stats
links
```

```python
data = requests.get(f"https://fbref.com{links[0]}")         #can call data.text
shooting = pd.read_html(data.text, match="Shooting")[0]
shooting.head()
```

```python
#combining tables
team_data = matches[0].merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
team_data.head()
```
